```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import astropy.time as astt
import datetime as dt
```

```python
data = ascii.read ('/Users/dharbeck/tfndomeb_shutterresponsetime.csv')
```

```python
 #data['dateobs'] = t['dateobs'].astype(str)
 data['dateobs'] = astt.Time(data['from_unixtime(s.TIMESTAMP_/1000)'], scale='utc', format=None).to_datetime()
 my_day = dt.date(2025, 6, 15)
 data['time_of_day'] = [ dt.datetime.combine(my_day, t.time()) for t in data['dateobs']]

```

```python
from matplotlib.pyplot import figure


plt.figure(figsize=(8, 6), dpi=200)
plt.plot (data['dateobs'],data['VALUE_'], ',')
plt.savefig("delay_dataobs.png", format='png', bbox_inches='tight')


plt.figure(figsize=(8, 6), dpi=200)

plt.plot (data['time_of_day'],data['VALUE_'], ',')

extreme = data['VALUE_']>1000
plt.plot (data['time_of_day'][extreme],data['VALUE_'][extreme], '.')

plt.xlabel('Time of Day')
plt.ylabel('Shutter Response Time (ms)')
plt.title('Shutter Response Time vs Time of Day')
plt.savefig("delay_timeofday.png", format='png', bbox_inches='tight')

```
